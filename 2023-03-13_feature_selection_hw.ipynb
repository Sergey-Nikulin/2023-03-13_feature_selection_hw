```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
# from sklearn.metrics import make_scorer #recall_score, precision_score, f1_score
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.datasets import make_classification
x_data_generated, y_data_generated = make_classification(scale=1, random_state = 0)
```

```python
X = x_data_generated
y = y_data_generated
# len(X), len(y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(random_state = 0)
model
```

```python
model.fit( X, y )
predictions = model.predict_proba( X )
```

```python
cross_val_score(model, X, y, scoring='accuracy').mean()
```

```python
df = pd.DataFrame(X)
df.describe()
```

```python
print('Матрица корреляции')
df_cor = df.corr()
df_cor[df_cor > 0.6]# выделяем признаки с коэф.кор.> 0.6
```

```python
# методом подстановки получаем ручший результат при удалении признака 17
df = df.drop(df.columns[[17]], axis=1)
df
```

```python
model = LogisticRegression(random_state = 0)
model.fit( df, y )
```

```python
cross_val_score(model, df, y, scoring='accuracy').mean()
# результат улучшился
```

```python
###
```

```python
 from sklearn.feature_selection import VarianceThreshold
```

```python
selector = VarianceThreshold(threshold=0.9)
dfs = selector.fit_transform(df)
dfs = pd.DataFrame(dfs)
dfs
```

```python
model = LogisticRegression(random_state = 0)
model.fit( dfs, y )
```

```python
cross_val_score(model, dfs, y, scoring='accuracy').mean() # результат улучшился
```

```python
###
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
X_new = SelectKBest(f_classif, k=5).fit_transform(dfs, y)
```

```python
dfs.shape, X_new.shape
```

```python
model = LogisticRegression(random_state = 0)
model.fit( X_new, y )
cross_val_score(model, X_new, y, scoring='accuracy').mean() # результат улучшился
```

```python
###
```

```python
from sklearn.feature_selection import SelectFromModel
# from sklearn.linear_model import LogisticRegression
```

```python
selector = SelectFromModel(estimator=LogisticRegression(penalty='l1', random_state = 0, solver='liblinear')).fit(X_new, y)
```

```python
selector.estimator_.coef_
```

```python
selector.threshold_
```

```python
selector.get_support()
```

```python
sfm = selector.transform(X_new)
sfm = pd.DataFrame(sfm)
sfm
```

```python
model = LogisticRegression(random_state = 0)
model.fit( sfm, y )
cross_val_score(model, sfm, y, scoring='accuracy').mean() # результат незначительно ухудшился
```

```python
###
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf_rf = RandomForestClassifier(max_depth=2, random_state=0)
clf_rf.fit(sfm, y)
```

```python
imp = pd.Series(clf_rf.feature_importances_)
```

```python
imp.index = pd.Series(sfm.columns)
imp = imp.sort_values(ascending=False)
imp
```

```python
sfm_n = sfm.drop(sfm.columns[[3]], axis=1)
sfm_n
```

```python
model = LogisticRegression(random_state = 0)
model.fit( sfm_n, y )
cross_val_score(model, sfm_n, y, scoring='accuracy').mean() # результат улучшился
```

```python
###
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
sfs = SequentialFeatureSelector(knn, n_features_to_select=2)
sfs.fit(sfm_n, y)
```

```python
sfs.get_support()
```

```python
sfs.transform(sfm_n).shape
```

```python
sfs_knn = sfs.transform(sfm_n)
sfs_knn = pd.DataFrame(sfs_knn)
sfs_knn
```

```python
model = LogisticRegression(random_state = 0)
model.fit( sfs_knn, y )
cross_val_score(model, sfs_knn, y, scoring='accuracy').mean() # результат не изменился
```
